### 2D Line plot

image.png

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# plotting a simple function
price = [48000,54000,52000,49000,48000,52000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)  # plt.plot(x,y)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#F55C3B')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#535955')

plt.title('Rohit Sharma Vs Virat Kohli IPL Batting Comparison')  # Graph's Title
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#F55C3B',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#535955',linestyle='dashdot', linewidth='3')

plt.title('Rohit Sharma Vs Virat Kohli IPL Batting Comparison')  # Graph's Title
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#F55C3B',marker='o', markersize='10')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#535955', linewidth='3',marker='D')

plt.title('Rohit Sharma Vs Virat Kohli IPL Batting Comparison')  # Graph's Title
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#F55C3B',marker='o', markersize='7', label ='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#535955', linewidth='3',marker='D', label = 'Rohit')

plt.title('Rohit Sharma Vs Virat Kohli IPL Batting Comparison')  # Graph's Title
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper left') # For label
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2021)
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='#F55C3B',marker='o', markersize='7', label ='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#535955', linewidth='3',marker='D', label = 'Rohit')

plt.title('Rohit Sharma Vs Virat Kohli IPL Batting Comparison')  # Graph's Title
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show -> to print the graph
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()

plt.show()
```

image.png

- Bivariate Analysis

- Use case - Finding correlation

```python
# plt.scatter simple function
x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,100,50)
y
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data
df = pd.read_csv('batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'],df['strike_rate'],color='olive',marker='o')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# marker
```

```python
# size
tips = sns.load_dataset('tips')


# slower
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*10)
```

```python
# scatterplot using plt.plot
# faster
plt.plot(tips['total_bill'],tips['tip'],'o')
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='navy')
```

```python
# bar chart using data
```

```python
# horizontal bar chart
plt.barh(colors,children,color='black')
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.15)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.15)
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.15)

plt.xticks(np.arange(df.shape[0]), df['batsman'])

plt.show()
```

```python
np.arange(df.shape[0])
```

```python
# Multiple Bar charts
```

```python
# xticks
```

```python
# a problem
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# Stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']),label='2015')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('/content/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','pink','cyan','brown'])
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1])
plt.show()
```
