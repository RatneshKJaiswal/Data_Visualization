```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('dark_background')
```

### Colored Scatterplots

```python
iris = pd.read_csv('iris.csv')
iris.shape
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='GnBu',alpha=0.5)  # c is colour according to, alpha is opacity, cmap is colourmap
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# cmap and alpha
```

### Plot size

```python
# Resizing the Graph

plt.figure(figsize=(13,6))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='GnBu',alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Annotations

```python
batters = pd.read_csv('batter.csv')
```

```python
batters.shape
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(16,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])    # size of point is based on  runs

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i]+0.6,sample_df['strike_rate'].values[i],sample_df['batter'].values[i],fontdict={'size':8,'color':'pink'})
```

```python
x = [1,2.5,3,4]
y = [5,6,8,4]

plt.scatter(x,y)
plt.text(1,5,'Point 1')
plt.text(2.5,6,'Point 2')
plt.text(3,8,'Point 3')
plt.text(4,4,'Point 4',fontdict={'size':14,'color':'cyan'})
```

### Horizontal and Vertical lines

```python
plt.figure(figsize=(14,6))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])    # size of point is based on  runs

plt.axvline(sample_df['avg'].mean(),color='red')
plt.axhline(sample_df['strike_rate'].mean(),color='pink')
plt.axhline(140,color='red')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i]+0.6,sample_df['strike_rate'].values[i],sample_df['batter'].values[i],fontdict={'size':8,'color':'green'})
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red',marker='.')
ax[1].scatter(batters['avg'],batters['runs'],marker='.')

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,6))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red',marker='.')
ax[0,1].scatter(batters['avg'],batters['runs'],marker='.')
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])

fig.show()
```

### 3D Surface Plots

```python
# z = x^2 + Y^2

x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='magma')
fig.colorbar(p)
```

```python
z = 1/(15*(xx**2 + yy**2))
# 1/(15(x^2+y^2))

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx)*yy

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)


fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Contour Plots

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
z = np.sin(xx) + np.cos(yy)
p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()
z = np.sin(xx) + np.cos(yy)
p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + yy

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Heatmap

```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head(20)
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
```

```python
# pie -> single and multiple
df = pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[2,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df.head()
```

```python
df['match1'].plot(kind='pie',labels=df['batsman'].values,autopct='%0.1f%%', textprops={'color':"Green"})
```

```python
# multiple pie charts

df[['match1','match2','match3']].plot(kind='pie',subplots=True,figsize=(15,8), labels=df['batsman'].values,autopct='%0.1f%%', textprops={'color':"Green"})
```
