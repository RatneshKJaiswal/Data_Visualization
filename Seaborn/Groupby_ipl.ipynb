```python
import pandas as pd
import numpy as np
```

```python
dileverys = pd.read_csv('deliveries.csv')
dileverys.head(10)
```

```python
runs = dileverys.groupby('batsman')
```

```python
runs['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
six = dileverys['batsman_runs'] == 4
new_deliveries = dileverys[six]
new_deliveries.shape
```

```python
new_deliveries.groupby('batting_team')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

```python
ms = dileverys[dileverys['batsman'] == 'MS Dhoni']
ms.head()
```

```python
ms.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# function for the above code

def run_scored(batsman_name):
    bat = dileverys[dileverys['batsman'] == batsman_name]
    return bat.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
run_scored('MS Dhoni')
```

Merge  ->  using two or more **groups**

```python
match = pd.read_csv('matches.csv')
match
```

```python
new = dileverys.merge(match, left_on='match_id', right_on='id')
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season')[['season','batsman']]
```

# Pivot column


```python
dileverys
```

```python
mask = dileverys['batsman_runs'] == 6
six = dileverys[mask]
six.head()
```

```python
pt=six.pivot_table(index='over', columns='batting_team', values='batsman_runs', aggfunc='count')
```

```python
# now converting it into a het map
import seaborn as sns
```

```python
sns.heatmap(pt)
```
